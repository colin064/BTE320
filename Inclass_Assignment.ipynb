```python
sales = float(input('What are the Sales: '))
TR = float(input('What is the Tax Rate: ')) /100
COGS = sales * 0.4
GP = sales - COGS
taxes = GP * TR
net = GP - taxes
print(f'Sales: {sales}\nCost of Goods Sold: ${COGS}\nGross Profit: ${GP}\nTaxes: {taxes}\nNet Income: {net}')
```

```python
num1 = float(input("Enter first number "))
num2 = float(input("Enter second number "))
sign = (input("*,+,-,/ "))
if sign == "+":
  SUM = num1 + num2
  print(SUM)
if sign == "-":
  SUM = num1 - num2
  print(SUM)
if sign == "/":
  SUM = num1 / num2
  print(SUM)
if sign == "*":
  SUM = num1 * num2
  print(SUM)
```
