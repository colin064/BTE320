```python
sales = float(input('What are the Sales: '))
TR = float(input('What is the Tax Rate: ')) /100
COGS = sales * 0.4
GP = sales - COGS
taxes = GP * TR
net = GP - taxes
print(f'Sales: {sales}\nCost of Goods Sold: ${COGS}\nGross Profit: ${GP}\nTaxes: {taxes}\nNet Income: {net}')
```

```python
num1 = float(input("Enter first number "))
num2 = float(input("Enter second number "))
sign = (input("*,+,-,/ "))
if sign == "+":
  SUM = num1 + num2
  print(SUM)
if sign == "-":
  SUM = num1 - num2
  print(SUM)
if sign == "/":
  SUM = num1 / num2
  print(SUM)
if sign == "*":
  SUM = num1 * num2
  print(SUM)
```

```python
print("Hour Fee")
for hour in range(1,9):
  fee = 5 + 2.5 * hour
  if fee < 10:
    fee = 10
  if fee > 20:
    fee = 20
  print (f"{hour}    {fee}")
```
