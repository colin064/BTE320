```python
def getValues(n):
    assert isinstance(n, int), "n must be an integer"
    items = []

    for i in range(n):
        while True:
            try:
                items.append(float(input("Enter a value: ")))
                break
            except ValueError:
                print("Invalid input. Please enter a number.")

    if len(items) < 3:
        raise ValueError("The list must have at least 3 items.")

    return items

def calcAverage(alist):
    assert isinstance(alist, list), "alist must be a list"

    if len(alist) < 3:
        raise ValueError("The list must have at least 3 items to drop the first two.")

    return sum(alist[2:]) / len(alist[2:])

def main():
    try:
        n = int(input("How many numbers: "))
        alist = getValues(n)
        avg = calcAverage(alist)

        print(f"Average is: {avg:.2f} lbs.")

    except ValueError as e:
        print(f"Error: {e}")

if __name__ == "__main__":
    main()
```

```python
class Student:
  def __init__(self, name, number):
    self.__name=name
    self.number = number
    self.__courses = []

  def get_name(self):
     return self.__name

  def get_courses(self):
    return self.__courses

  def set_name(self, newName):
    self.__name = newName

  def __len__(self):
      return len(self.__courses)

  def enroll(self, newCourse):
    if newCourse not in self.__courses:
      self.__courses.append(newCourse)
    else:
      print(f"student {self.__name} has already enrolled in {newCourse}.")
```

```python
class Point:
  def __init__(self, x, y):
    self.x = x
    self.y = y

  def distance(self, other):
    dx = (self.x - other.x) ** 2
    dy = (self.y - other.y) ** 2
    return (dx + dy) ** 0.5

class Rectangle:
  def __init__(self, p1_coords, p2_coords):
    x1, y1 = p1_coords
    x2, y2 = p2_coords

    x_min = min(x1, x2)
    x_max = max(x1, x2)
    y_min = min(y1, y2)
    y_max = max(y1, y2)

    self.p_bottom_left = Point(x_min, y_min)
    self.p_bottom_right = Point(x_max, y_min)
    self.p_top_left = Point(x_min, y_max)
    self.p_top_right = Point(x_max, y_max)

  def width(self):
    return self.p_bottom_right.x - self.p_bottom_left.x

  def height(self):
    return self.p_top_left.y - self.p_bottom_left.y

  def area(self):
    return self.width() * self.height()
```

```python
coords_1 = [0,3]
coords_2 = [3,0]
z = Rectangle(coords_1, coords_2)
print(f"Area of the rectangle: {z.area()}")
```

```python
class Employee:
  def __init__(self,name,wage,hours):
    self.name = name
    self.__hourly_wage = wage
    self.hours = hours

  def get_wage(self):
    return self.__hourly_wage

  def set_wage(self, newWage):
    if newWage <= 0:
      print("Negative wage values not allowed")
    else:
      self.__hourly_wage = newWage

  def calculate_pay(self):
    return self.__hourly_wage * self.hours

  def __str__(self):
    return f"Employee {self.name}\nHourly Wage: {self.__hourly_wage}"

class Manager(Employee):
  def __init__(self,name,wage,hours,bonus):
    super().__init__(name,wage,hours)
    self.bonus = bonus

  def calculate_pay(self):
    return (1 + self.bonus) * self.get_wage() * self.hours

employee = Employee("Bob", 50, 10)
manager = Manager("Alice", 100,15,10)

print(employee)
print(manager)
print(f"Employee pay: {employee.calculate_pay()}")
print(f"Manager pay: {manager.calculate_pay()}")
```
